```python
import wandb
api = wandb.Api()
run = api.run("/vlm_rl/new_rl_ovon_full/runs/ai46tvj7")
episode_labels = set(run.history(samples=5000)['episode_label'])
```

```python
import json
ovon_val_synonym_full_set = json.load(open('/Projects/spatial_training/conf/episode_jsons/ovon_val_seen_synonyms.json','r'))
remaining_list = []
ovon_val_synonym_full_set.reverse()
for item in ovon_val_synonym_full_set:
    if item not in episode_labels:
        remaining_list.append(item)
```

```python
json.dump(remaining_list,open('remaining_episodes.json','w'))
```
